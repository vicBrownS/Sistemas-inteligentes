# Introducción a Git  

`Git` (Global Information Tracker) es un sistema de control de versiones que te permite rastrear cambios en tus archivos y colaborar con otros usuarios.

## Términos básicos de Git


- El `directorio Git` (ubicado en RUTA-DE-TU-PROYECTO/.git/) es donde Git almacena todo lo que necesita para hacer un seguimiento preciso del proyecto. Esto incluye metadatos y una base de datos de objetos que incluye versiones comprimidas de los archivos del proyecto.

- El `directorio de trabajo` es donde un usuario realiza cambios locales en un proyecto. El directorio de trabajo descarga los archivos del proyecto de la base de datos de objetos del directorio Git y los coloca en la máquina local del usuario.

### Remote

Cuando clonas un repositorio, **git** automáticamente configura un `remoto` llamado `origin` que apunta al repositorio desde donde se realizó la clonación.

Para ver los remotos asociados a un repositorio, usa el siguiente comando:

```bash
git remote -v
```

Puedes agregar otros remotos usando el siguiente comando:

```bash
git remote add <nombre> <URL_del_repositorio>
```

Por ejemplo:

```bash
git remote add origin2 https://github.com/usuario/repositorio.git
```

### Estados de Git

Cuando trabajas con Git, los archivos pueden encontrarse en tres estados diferentes:

1. **Modificado**: Has cambiado el archivo, pero aún no está en el área de staging.
2. **Preparado (Staged)**: El archivo ha sido marcado para ser incluido en el próximo commit.
3. **Confirmado (Committed)**: El archivo ha sido guardado en el historial del repositorio.

Los cambios en un repositorio se deben realizar de una determinada manera. El proceso principal para ello es el `commit`, que podemos traducir como consolidar o confirmar.
Cada vez que un archivo del repositorio local es modificado, es candidato a ser enviado al repositorio remoto. Antes de poder las modificaciones al remoto, debemos incluirlas en una especie de area de ensayo (staging area o stage). Este proceso se realiza mediante el comando git `add`. Veamos esto con un ejemplo:

En el repositorio que has clonado, crea una carpeta <kbd>code</kbd> y dentro incluye un archivo con nombre `programa.py` con el siguiente código:

```Python
# -*- coding: utf-8 -*-
"""
Created on Fecha y hora

@author: tu_nombre
"""

print("Programa nuevo")

````
De esta manera hemos realizado una modificación en el repositorio. 



#### Status
Para conocer que archivos han sufrido modificación o se han incluido en el repositorio:


```bash
# Ver el estado de los archivos
git status
```
El resultado de esta instrucción nos informará que el archivo `code/programa_nuevo.py` está sin seguimiento. 

#### Add
Para incluir el archivo en el area de stage, ejecutaremos el siguiente código en la terminal:

```bash
# Añadir un archivo al stage
git add code/programa_nuevo.py

# Agregar múltiples archivos al área de staging.
git add <archivo1> <archivo2>

# Añadir todos los archivos al stage
git add .
```

Si comprobamos de nuevo el estado del repositorio, nos informará que existen cambios que faltan por ser confirmados. 


#### Reset
Para retirar un archivo del area de stage, que hayamos incluido erroneamente:


```bash
# Añadir un archivo al stage
git reset <nombre_del_archivo>
```
El comando retirará el archivo indicado del área de staging, pero se mantendrán los cambios realizados en el archivo, pudiendo editarse y modificarse antes de volver a agregarlo al área de staging.

#### Commit

El proceso de *commit* o consolidación, lo realizaremos mediante el siguiente comando, en el cual incluimos un mensaje que nos informa de los procesos que componen la consolidación:

```bash

git commit -m "Añadimos el archivo programa_nuevo.py en la carpeta code"

```

Este mensaje es importante que sea breve pero que a la vez contenga la información necesaria para conocer las modificaciones o actualizaciones que van a producir en el repositorio.


<img style="float:center" width="60%" src="pics/commit_001.png">



Cada vez que realicemos modificaciones en archivos o creemos algún nuevo fichero y consideremos que forman parte de un proceso específico, tendremos que realizar todo el proceso anterior (añadir mediante `git add` y consolidar con `git commit`)

Podemos obtener un historial de commits mediante el siguiente comando:
```bash
git log
```

### Subir los cambios al repositorio remoto
Una vez que tenemos uno o varios commits, podemos subir estos cambios al repositorio remoto empleando el siguiente comando, indicando el repositorio remoto `origin` y la rama en la que nos encontramos trabajando (en nuestro caso aun no hemos generado ninguna otra rama y la única de la que disponemos es `main`):

```bash
git push origin main
```

## Control de archivos a subir al repositorio

El repositorio de ejemplo se encuentra un archivo con nombre `.gitignore`. Este fichero nos va a permitir especificar qué archivos o directorios deben ser ignorados por Git. Esto es útil para evitar el seguimiento de archivos temporales, de configuración local, o cualquier otro archivo que no deba ser incluido en el control de versiones. 


#### ¿Qué es el Archivo `.gitignore`?

El archivo `.gitignore` es un archivo de texto simple que contiene una lista de archivos o directorios que Git debe ignorar. Cada línea en el archivo `.gitignore` representa un patrón que coincide con los nombres de archivos o directorios a ser ignorados.

---

#### Configuración de un Archivo `.gitignore`

- 1: Crear el archivo en el repositorio local.
- 2: Añadir Patrones al Archivo
    - Abre el archivo `.gitignore` en un editor de texto y añade los patrones de los archivos o directorios que deseas ignorar. Por ejemplo:
    
```
# Ignorar archivos de sistema operativo:
.DS_Store
```

---

#### Patrones Básicos en `.gitignore`

- Ignorar Archivos por Extensión: Para ignorar todos los archivos de una cierta extensión, usa el carácter `*`:

```
*.log   # Ignora todos los archivos .log
*.tmp   # Ignora todos los archivos .tmp
```

- Directorios Completos: Para ignorar un directorio completo, añade una barra `/` al final del nombre del directorio:
```
logs/   # Ignora el directorio logs y todo su contenido
tmp/    # Ignora el directorio tmp y todo su contenido
```

- Ignorar Archivos o Directorios Específicos: Para ignorar archivos o directorios específicos, simplemente añade sus nombres:
```
secret.txt   # Ignora el archivo secret.txt
build/       # Ignora el directorio build y todo su contenido
```

#### Ejemplos de `.gitignore`

- Ignorar Archivos de Sistema Operativo

```
# Ignorar Archivos de Sistema Operativo
# OsX
.DS_Store

# Windows
Thumbs.db
desktop.ini

# Ignorar archivos y directorios de compilación
build/
dist/
*.class
*.o
*.so
*.exe

# Ignorar dependencias de Python
__pycache__/
*.py[cod]
.Python
env/
venv/
.venv/
```

#### Verificación de Archivos Ignorados

Para verificar qué archivos están siendo ignorados por Git, puedes usar el siguiente comando:
```bash
git status --ignored
```
Este comando mostrará una lista de archivos y directorios que están siendo ignorados según los patrones del archivo `.gitignore`.

### Gestión de ramas en GitHub

Las ramas (`branches`) en Git son una característica fundamental que permite a los desarrolladores trabajar en diferentes líneas de desarrollo dentro de un mismo repositorio. Las ramas son muy útiles para desarrollar nuevas funcionalidades, corregir errores o experimentar sin afectar la rama principal (`main` o `master`).


Algunos conceptos fundamentales:

- **Rama (Branch)**: Es una versión del repositorio que diverge de la línea principal. Permite desarrollar funcionalidades de forma aislada.
- **Rama Principal (Main/Master)**: Es la rama por defecto donde se encuentra el código fuente estable.
- **Fusión (Merge)**: Es el proceso de integrar los cambios de una rama en otra.
- **Conflictos**: Ocurren cuando hay cambios incompatibles en las ramas que se están fusionando. Requieren resolución manual.

#### Crear una Nueva Rama

Para crear una nueva rama y cambiar a ella inmediatamente, utiliza el siguiente comando:

```bash
git checkout -b nombre-de-la-rama
```

Este comando hace dos cosas:
1. Crea una nueva rama llamada `nombre-de-la-rama`.
2. Cambia a la nueva rama.

#### Cambiar a una rama existente
Para cambiar a una rama que ya existe, utiliza:

```bash
git checkout nombre-de-la-rama
```

De esta forma, los `commit` que se realicen, afectarán unicamente a la rama activa.


Podéis incluir un archivo de imagen en la carpeta `images` y a continuación:
- `add`
- `commit`
- `push` (en la rama *nombre-de-la-rama*)

Si compruebas como está tu repositorio remoto, encontrarás que se ha creado una nueva rama y que la nueva imagen si que aparece en esa rama, pero no en la rama `main`.


#### Listar Todas las Ramas

Para ver todas las ramas disponibles en el repositorio, utiliza:

```bash
git branch
```
En este caso, aparecerán dos ramas (`main` y `nombre-de-la-rama`) marcando de manera específica aquella en la que se encuentra trabajando actualmente (aparece marcada con un asterisco (*)).

#### Verificar la Rama Actual

Para verificar en qué rama estás actualmente, puedes emplear de nuevo el comando `status`:

```bash
git status
```

#### Fusionar una Rama en Otra

Para fusionar los cambios de una rama en otra, para lo que disponemos del comando `merge`:

1. Cambia a la rama donde quieres fusionar los cambios:
    ```bash
    git checkout main
    ```
2. Fusiona los cambios de la otra rama:
    ```bash
    git merge nombre-de-la-rama
    ```

#### Resolver conflictos de fusión

Si hay conflictos durante la fusión, Git te lo notificará y tendrás que resolverlos manualmente:

1. Abre los archivos en conflicto y resuélvelos.
2. Añade los archivos resueltos al área de preparación:
    ```bash
    git add archivo-en-conflicto.txt
    ```
3. Completa la fusión:
    ```bash
    git commit -m "Resolver conflictos de fusión"
    ```

#### Eliminar una Rama Localmente
Para eliminar una rama que ya no necesitas localmente, usa:

```bash
git branch -d nombre-de-la-rama
```

Si la rama no se ha fusionado completamente, puedes forzar la eliminación usando:

```bash
git branch -D nombre-de-la-rama
```

#### Eliminar una Rama en el Repositorio Remoto
Para eliminar una rama en el repositorio remoto, usa:

```bash
git push origin --delete nombre-de-la-rama
```


### Mantener el repositorio local actualizado

Si estamos trabajando de forma colaborativa, es posible que se realicen modificaciones del repositorio remoto en la rama en la que estamos desarrollando y que tengamos que actualizar en nuestro repositorio local.
Para ello, tendremos que realizar los siguientes procesos:

- `git fetch`: permite descargar commits, archivos y referencias de un repositorio remoto al repositorio local. Este proceso no fusiona los cambios automáticamente en la rama actual, con la oportunidad de revisar los cambios antes de integrarlos.

Obtener los cambios del repositorio remoto:
```bash
git checkout nombre-de-la-rama
git fetch origin
```
Después de hacer un fetch, puedes ver las nuevas ramas descargadas utilizando:
```bash
git branch -r
```
Este comando lista todas las ramas remotas disponibles.


Una vez revisados los cambios, puedes fusionar los cambios en la rama principal local (main):

```bash
    git checkout main
    git merge nombre-de-la-rama
```


- `git pull` es un comando que combina `git fetch` y `git merge`. Descarga los cambios del repositorio remoto y los fusiona automáticamente con tu rama actual.

Para obtener y fusionar los últimos cambios del repositorio remoto en tu rama actual, usa el comando:
```bash
git pull origin main
```
Esto descargará y fusionará los cambios de la rama `main` del remoto `origin` en tu rama actual.

##### Diferencias entre `git fetch` y `git pull`

- `git fetch` solo descarga los cambios del repositorio remoto y los almacena en tu repositorio local.
- `git pull` descarga los cambios y los fusiona automáticamente con tu rama actual.

Un **issue** es una herramienta para rastrear tareas, mejoras, errores, preguntas y más dentro de un proyecto. Es una manera estructurada de gestionar el trabajo pendiente y realizar un seguimiento de las incidencias que surgen durante el desarrollo.

3. Introduce un título y una descripción detallada.

5. Etiqueta el issue con un label apropiado y asigna un milestone si corresponde.
6. Haz clic en **Submit new issue** para crear el issue.

### `Labels`

Los `labels` son etiquetas que se utilizan para categorizar los **issues**. Permiten identificar rápidamente el tipo de trabajo que requiere un **issue** y priorizar el trabajo.

#### Pasos para emplear/crear un `label`

1. En la pestaña `Issues` de tu repositorio, selecciona un **issue** existente o crea uno nuevo.
2. En la barra lateral derecha, selecciona `Labels`.
3. Elige un **label existente** o crea uno nuevo.


##### Ejemplo de `Labels` comunes

- `bug`: Indica un error que debe ser corregido.
- `enhancement`: Para mejoras o nuevas características.
- `question`: Para preguntas o solicitudes de información.
- `documentation`: Relacionado con la documentación del proyecto.
- `wontfix`: Para issues que no se planean resolver.
